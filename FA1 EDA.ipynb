```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
```

```python
df = pd.read_csv("cytof_one_experiment.csv")
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['X2B4'], bins=30, kde=False, color='skyblue', edgecolor='black')
plt.title('Histogram of X2B4')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(df['X2B4'], color='green', fill=True)
plt.title('Density Estimate Plot of X2B4')
plt.xlabel('Values')
plt.ylabel('Density')
plt.grid(True)
plt.show()
```

# The column of X2B4 is a right skewed. the data is concentrated more on one side

```python
X2B4_sorted = np.sort(df['X2B4'])
CD69_sorted = np.sort(df['CD69'])
```

```python
min_length = min(len(df['X2B4']), len(df['CD69']))
norm_quantiles = stats.norm.ppf(np.linspace(0.01, 0.99, min_length))
```

```python
plt.figure(figsize=(10, 10))
plt.scatter(norm_quantiles, X2B4_sorted, label='X2B4', alpha=0.5)
plt.scatter(norm_quantiles, CD69_sorted, label='CD69 ', alpha=0.5)
plt.plot(norm_quantiles, norm_quantiles, color='red', linestyle='--', label='Expected')
plt.title('Q-Q Plot of X2B4 vs CD69')
plt.xlabel('Theoretical Quantiles')
plt.ylabel('Ordered Values')
plt.legend()
plt.grid(True)
plt.show()
```

#Points above the diagonal line suggest that the values in one column tend to be higher than the corresponding values in the other column.
